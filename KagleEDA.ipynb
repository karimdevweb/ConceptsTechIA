- since we have a list of files, i ddecided to read them directly from zip and creae correspondent variables

```python
zip_path = 'Data/home-credit-default-risk.zip'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    for file_name in zip_ref.namelist():
        if file_name.lower().endswith('.csv'):
            try:
                with zip_ref.open(file_name) as f:
                    # Create a variable name based on the file name
                    base_name = os.path.splitext(os.path.basename(file_name))[0]  # remove path and extension
                    var_name = f"{base_name}_df"
                    
                    # Dynamically create the variable
                    globals()[var_name] = pd.read_csv(f, encoding='cp1252')
                    
                    print(f"Created variable: {var_name}")
            except Exception as e:
                print(f"Failed to read {file_name}: {e}")
```
